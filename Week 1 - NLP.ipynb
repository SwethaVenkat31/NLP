## 1.1 Simple Frequency Distribution


```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
#Initialize tokenizer
tokenizer=TreebankWordTokenizer()

#sample text
text="Natural Language Processing (NLP) with python is fun and educational . Python simplifies NLP tasks."

#tokenize text
tokens=tokenizer.tokenize(text)

#print tokens
print("Tokens:",tokens)

#frequency distribution
fdist=FreqDist(tokens)

#print frequency
print("\n Word Frequency Distribution:")
for word, freq in fdist.items():
    print(f"{word}:{freq}")

#plot frquency distribution 
plt.figure(figsize=(15,8))
plt.hist(tokens,bins=5)
plt.show()
```

```python
fdist.plot(title='Simple Frequency Distribution plot')
plt.show()
```

## 1.2 Advanced Frequency Distribution with stopwords Removal

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
nltk.download('stopwords')
```

```python
#Initialize tokenizer
tokenizer=TreebankWordTokenizer()
stop_words=set(stopwords.words('english'))

#sample text
text= "Natural Language Processing with python is interesting and useful. Python makes NLP tasks easier and more efficient."

#tokenize
tokens=tokenizer.tokenize(text)

#Removal stopwords
filtered_tokens=[word for word in tokens if word.lower() not in stop_words and word.isalpha()]

#print filtered tokens
print("Filtered tokens(withour stopewords)",filtered_tokens)
fdist=FreqDist(filtered_tokens)
print("\n Word Frequency Distribution (without stopwords):")

for word,freq in fdist.items():
    print(f"{word}:{freq}")
fdist.plot(title='Word frquency distribution(without stopwords)')
plt.show()
```

```python
plt.hist(filtered_tokens,bins=15)
plt.show()
```
